```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a sample sales network graph
sales_network = nx.Graph()

# Add nodes for products and stores
sales_network.add_nodes_from(["Product A", "Product B", "Store 1", "Store 2", "Store 3"])

# Add edges representing sales connections
sales_network.add_edges_from([("Product A", "Store 1"), ("Product A", "Store 2"),
                               ("Product B", "Store 2"), ("Product B", "Store 3"),
                               ("Store 1", "Store 2"), ("Store 2", "Store 3")])

# Define the force layout
pos = nx.spring_layout(sales_network)

# Draw the sales network graph
nx.draw(sales_network, pos, with_labels=True, node_size=1000, node_color='skyblue', font_size=10)

# Show the graph
plt.show()
```
